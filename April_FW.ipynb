```python
#1-1
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

read = pd.read_csv('/content/drive/MyDrive/머신러닝/data0405.csv')

x = read.iloc[:,0].values
y = read.iloc[:,1].values
y = y.reshape(-1,1)
m = x.shape[0]
A = np.hstack([np.ones([m,1]), x.reshape(-1,1)])
A = np.asmatrix(A)
w = (A.T *  A).I*A.T*y

plt.figure(figsize = (6,4))
plt.title('Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(x,y,'ro')

xp = np.arange(min(x),max(x),0.01).reshape(-1,1) #x의 최솟값, 최댓값
yp = w[0,0] + w[1,0] * xp
plt.plot(xp,yp,'b-')
plt.show()
```

```python
#2-1
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

read = pd.read_csv('/content/drive/MyDrive/머신러닝/data0405.csv')

x = read.iloc[:,0].values
y = read.iloc[:,1].values

w1 = 0
w0 = 0
alpha = 0.001
epochs = 5000

for i in range(epochs) :
  y_pred = w1 * x + w0
  dw1 = (x*(y_pred - y)).mean()
  dw0 = (y_pred - y).mean()
  w1 = w1 - alpha * dw1
  w0 = w0 - alpha * dw0

print(w1, w0)

plt.figure(figsize = (6,4))
plt.title('Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(x,y,'ro')

xp = np.arange(min(x), max(x), 0.01)
yp = w1 * xp + w0
plt.plot(xp, yp, 'b-')
plt.show()
```

```python
#2-2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

read = pd.read_csv('/content/drive/MyDrive/머신러닝/data0405.csv')

x = read.iloc[:,0].values
y = read.iloc[:,1].values

K = 2
alpha = 0.001
epochs = 5000
x_matrix = np.zeros((len(x), K+1))
for i in range(K+1) :
  x_matrix[:,i] = x**i
xm_tp = x_matrix.T

weights_matrix = np.zeros(K+1)
for i in range(K+1) :
  weights_matrix[i] = np.random.uniform(-1,1)
wm_grad = np.zeros(K+1)
for j in range(epochs) :
  y_pred = np.dot(weights_matrix,xm_tp)
  error = y_pred - y
  for i in range(K+1) :
    wm_grad[i] = (error * xm_tp[i,:]).mean()
  weights_matrix = weights_matrix - alpha * wm_grad
print(weights_matrix)

plt.plot(x,y,'ro')

xp = np.arange(min(x), max(x), 0.01)
xp_matrix = np.zeros((len(xp), K+1))
for i in range(K+1) :
  xp_matrix[:,i] = xp**i
xp_tp = xp_matrix.T
yp = np.dot(weights_matrix, xp_tp)
plt.plot(xp, yp, 'b-')
plt.show()
```

```python
#3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

read = pd.read_csv('/content/drive/MyDrive/머신러닝/data0405.csv')

x1_2 = read.iloc[:,0].values
y1_2 = read.iloc[:,1].values
x1_2 = x1_2.reshape(-1,1)
y1_2 = y1_2.reshape(-1,1)
mine = 2
A = np.hstack([x1_2**i for i in range(mine + 1)])
A = np.asmatrix(A)
w = (A.T *  A).I * A.T * y1_2

plt.figure(figsize = (6,4))
plt.title('Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(x1_2,y1_2,'ro')

x1_2p = np.arange(np.min(x1_2),np.max(x1_2),0.01).reshape(-1,1) #x의 최솟값, 최댓값
y1_2p = 0

for i in range(mine + 1) :
   y1_2p += w[i,0] * x1_2p**i
plt.plot(x1_2p,y1_2p,'b-')

print('최소 제곱법으로 구한 파라미터 :')
for i in range(mine + 1) :
  print(f'{w[i, 0]}')

#--- --- --- ---#

x2_2 = read.iloc[:,0].values
y2_2 = read.iloc[:,1].values

K = 2
alpha = 0.001
epochs = 5000

x2_2_matrix = np.zeros((len(x2_2), K+1))
for i in range(K+1) :
  x2_2_matrix[:,i] = x2_2**i
x2_2m_tp = x2_2_matrix.T
y2_2 = y2_2.flatten()

weights_matrix = np.zeros(K+1)
for i in range(K+1) :
  weights_matrix[i] = np.random.uniform(-1,1)

wm_grad = np.zeros(K+1)
for j in range(epochs) :
  y_pred = np.dot(weights_matrix,x2_2m_tp)
  error = y_pred - y2_2
  for i in range(K+1) :
    wm_grad[i] = (error * x2_2m_tp[i,:]).mean()
  weights_matrix = weights_matrix - alpha * wm_grad

print('\n경사하강법으로 구한 파라미터 :')
for i in range(K+1) :
  print(f'{weights_matrix[i]}')

x2_2p = np.arange(np.min(x2_2), np.max(x2_2), 0.01).reshape(-1,1)
x2_2p_matrix = np.zeros((len(x2_2p), K+1))
for i in range(K+1) :
  x2_2p_matrix[:, i] = x2_2p.flatten() ** i

x2_2p_tp = x2_2p_matrix.T
y2_2p = np.dot(weights_matrix, x2_2p_tp)
plt.plot(x2_2p, y2_2p, 'm-')
plt.show()
```
